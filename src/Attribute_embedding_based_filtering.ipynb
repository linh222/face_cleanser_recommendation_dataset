```python
from elasticsearch import Elasticsearch
import pandas as pd
import numpy as np
import json
import requests
es = Elasticsearch(hosts=['http://localhost:9200'], timeout=100)
```

```python
df_ner = pd.read_csv('dataset/attribute_based_embeddings/attribute_text_filtering.csv')
```

```python
df_ner_embedding = pd.read_json('dataset/attribute_based_embeddings/attribute_embedding_filtering.json', orient='index')
df_ner_embedding
```

```python
list_used_col = ['product_name', 'ingredient', 'feature', 'skin_type', 'capacity', 'design',
                'brand', 'expiry', 'origin']
boost_score = {
    'product_name': 1,
    'ingredient': 1,
    'feature': 1,
    'skin_type': 1,
    'capacity': 1,
    'design': 0.001,
    'brand': 1,
    'expiry': 0.001,
    'origin': 1
}
```

```python
def recommend_product(product_id, boost_score):
    must = []
    data = df_ner[df_ner['product_id']==product_id]
    for col in list_used_col:

        if pd.isnull(data[col].values[0]) == False:

            must.append( {
                "field": col,
                "query_vector": df_ner_embedding[df_ner_embedding['product_id']==product_id][col].values[0],
                "k": 101,
                "num_candidates": 1000,
                "boost": boost_score[col],

            })

    query_template = {
            'knn': must,
        'size': 101,
        '_source': ['product_id', 'num_sold_time'],
    }

    query_template = json.dumps(query_template)
    url = 'http://localhost:9200' + '/_search'
    headers = {"Content-Type": "application/json"}
    result = requests.post(url, data=query_template, headers=headers).json()
    ranked_list = {}

    for i in range(0, len(result['hits']['hits'])):
        product_id = int(result['hits']['hits'][i]['_id'])
        ranked_list[product_id] = {'score': result['hits']['hits'][i]['_score'],
                                   'num_sold': result['hits']['hits'][i]['_source']['num_sold_time']}

        ranked_list[product_id]['score'] = ranked_list[product_id]['score']*0.7 + ranked_list[product_id]['num_sold']*0.3

    ranked_list = dict(sorted(ranked_list.items(), key=lambda x: x[1]['score'], reverse=True))

    return ranked_list
```

```python
def NDCG(y_true, y_pred, k=None):
    if y_true not in y_pred:
        return 0.0

    true_index = y_pred.index(y_true)
    ideal_sorted = sorted(y_pred, reverse=True)
    ideal_index = ideal_sorted.index(y_true)

    dcg = 1.0 / np.log2(true_index + 2)
    idcg = 1.0 / np.log2(ideal_index + 2)

    ndcg = dcg / idcg

    if k is not None:
        ideal_sorted = ideal_sorted[:k]
        y_pred = y_pred[:k]
        true_index = y_pred.index(y_true)
        ideal_index = ideal_sorted.index(y_true)
        dcg = 1.0 / np.log2(true_index + 2)
        idcg = 1.0 / np.log2(ideal_index + 2)
        ndcg = dcg / idcg

    return ndcg
def MRR(y_true, y_pred):
    if y_true not in y_pred:
        return 0

    rank = 0
    for index, value in enumerate(y_pred):
        if value == y_true:
            rank = 1 / (index + 1)
            break

    return rank
```

```python
def PtopK(y_true, y_pred):
  if y_true in y_pred:
    return 1/len(y_pred)
  else:
    return 0
def RtopK(y_true, y_pred):
  if y_true in y_pred:
    return 1
  else:
    return 0
```

# Perform test

```python
df_purchase = pd.read_csv('/dataset/data_purchase.csv')
df_purchase
```

```python
rtopk_val = []
rtopk_test = []
ptopk_val = []
ptopk_test = []
mrr_val = []
mrr_test = []
ndcg_val = []
ndcg_test = []

for user in df_purchase['user_id'].unique():
  temp = df_purchase[df_purchase['user_id']==user]
  x_val = temp.iloc[2, 1]
  y_val = temp.iloc[1, 1]
  y_val_pred = recommend_product(product_id=x_val.tolist(), boost_score=boost_score)
  if x_val.tolist() in y_val_pred.keys():
      y_val_pred.pop(x_val.tolist())

  y_val_pred = list(y_val_pred.keys())[:10]
  rtopk_val.append(RtopK(y_val, y_val_pred))
  ptopk_val.append(PtopK(y_val, y_val_pred))
  mrr_val.append(MRR(y_val, y_val_pred))
  ndcg_val.append(NDCG(y_val, y_val_pred, 10))

  x_test = temp.iloc[1, 1]
  y_test = temp.iloc[0, 1]
  y_test_pred = recommend_product(product_id=x_test.tolist(), boost_score=boost_score)
  if x_test.tolist() in y_test_pred.keys():
      y_test_pred.pop(x_test.tolist())
  else:
      y_test_pred.pop(y_test_pred.keys()[-1])
  y_test_pred = list(y_test_pred.keys())[:10]
  rtopk_test.append(RtopK(y_test, y_test_pred))
  ptopk_test.append(PtopK(y_test, y_test_pred))
  mrr_test.append(MRR(y_test, y_test_pred))
  ndcg_test.append(NDCG(y_test, y_test_pred, 10))
```

```python
def calculate_mean(numbers):
    if not numbers:
        return 0.0
    return sum(numbers) / len(numbers)

print('rtopk val', calculate_mean(rtopk_val))
print('rtopk test', calculate_mean(rtopk_test))
print('ptopk val',calculate_mean(ptopk_val))
print('ptopk test', calculate_mean(ptopk_test))
print('mrr val', calculate_mean(mrr_val))
print('mrr test', calculate_mean(mrr_test))
print('ndcg val', calculate_mean(ndcg_val))
print('ndcg test', calculate_mean(ndcg_test))
```

# Grid search for weight parameters

```python
name_param = [0.001, 0.5, 1]
ingredient_feature_param = [0.001, 0.5, 1]
other_param = [0.001, 0.5, 1]
```

```python
import time
from tqdm import tqdm
```

```python
df = pd.DataFrame(columns = ['name_weight', 'ingredient_weight', 'other_weight', 'rtopk val', 'rtopk test', 'ptopk val', 'ptopk test', 'mrr val', 'mrr test', 'ndcg val', 'ndcg test'])
index = 0
for name_weight in tqdm(name_param):
    for ingredient_weight in ingredient_feature_param:
        for other_weight in other_param:
            start_time = time.time()
            boost_score = {
                        'product_name': name_weight,
                        'ingredient': ingredient_weight,
                        'feature': ingredient_weight,
                        'skin_type': other_weight,
                        'capacity': other_weight,
                        'design': other_weight,
                        'brand': other_weight,
                        'expiry': other_weight,
                        'origin': other_weight
                    }
            rtopk_val = []
            rtopk_test = []
            ptopk_val = []
            ptopk_test = []
            mrr_val = []
            mrr_test = []
            ndcg_val = []
            ndcg_test = []

            for user in df_purchase['user_id'].unique():
              temp = df_purchase[df_purchase['user_id']==user]
              x_val = temp.iloc[2, 1]
              y_val = temp.iloc[1, 1]
              y_val_pred = recommend_product(product_id=x_val.tolist(), boost_score=boost_score)
              if x_val.tolist() in y_val_pred.keys():
                  y_val_pred.pop(x_val.tolist())


              y_val_pred = list(y_val_pred.keys())[:10]
              rtopk_val.append(RtopK(y_val, y_val_pred))
              ptopk_val.append(PtopK(y_val, y_val_pred))
              mrr_val.append(MRR(y_val, y_val_pred))
              ndcg_val.append(NDCG(y_val, y_val_pred, 10))

              x_test = temp.iloc[1, 1]
              y_test = temp.iloc[0, 1]
              y_test_pred = recommend_product(product_id=x_test.tolist(), boost_score=boost_score)
              if x_test.tolist() in y_test_pred.keys():
                  y_test_pred.pop(x_test.tolist())
              else:
                  y_test_pred.pop(list(y_test_pred.keys())[-1])
              y_test_pred = list(y_test_pred.keys())[:10]
              rtopk_test.append(RtopK(y_test, y_test_pred))
              ptopk_test.append(PtopK(y_test, y_test_pred))
              mrr_test.append(MRR(y_test, y_test_pred))
              ndcg_test.append(NDCG(y_test, y_test_pred, 10))

            dict_temp = {
                'name_weight': name_weight,
                'ingredient_weight': ingredient_weight,
                'other_weight':other_weight,
                'rtopk val': calculate_mean(rtopk_val),
                'rtopk test': calculate_mean(rtopk_test),
                'ptopk val': calculate_mean(ptopk_val),
                'ptopk test': calculate_mean(ptopk_test),
                'mrr val': calculate_mean(mrr_val),
                'mrr test': calculate_mean(mrr_test),
                'ndcg val': calculate_mean(ndcg_val),
                'ndcg test': calculate_mean(ndcg_test)
            }
            df = pd.concat([df, pd.DataFrame(dict_temp, index=[index])], ignore_index=True)
            index += 1
            print(f'Total time for 1 run {name_weight}, {ingredient_weight}, {other_weight}: ', round(time.time()-start_time, 4))
```
